```python
def max_arrests(temp,borough):
  age_catagory=['<18','18-24','25-44','45-64','65+']
  max=temp[0]
  poss=0
  for i in range(1,len(temp)):
    if max<temp[i]:
       max=temp[i]
       poss=i
  print("Age group",age_catagory[poss],"from",borough,"maximum value")
```

Formula for converting values to radian is given by 
= value/57.29577951

Distance, d = 3963.0 * arccos[(sin(lat1) * sin(lat2)) + cos(lat1) * cos(lat2) * cos(long2  long1)]


```python

from math import radians, cos, sin, asin, sqrt,acos

def Dis_b_w_2_pnt(lat1,lon1,lat2,lon2):
  """The following function find the distance between 2 given set of coordinates"""
  lon1 = lon1/57.29577951
  lon2 = lon2/57.29577951
  lat1 = lat1/57.29577951
  lat2 = lat2/57.29577951
  d = 3963.0 * acos((sin(lat1) * sin(lat2)) + cos(lat1) * cos(lat2) * cos(lon2-lon1))
  return d

k=Dis_b_w_2_pnt(53.32055555555556,-1.7297222222222221,53.31861111111111, -1.6997222222222223)
print(k)
```
